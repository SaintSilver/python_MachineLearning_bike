# 바이크 수요예측

```python
import pandas as pd
import numpy as np
train = pd.read_csv("biketrain.csv", parse_dates = ["datetime"])
train.head()
```

```python
test = pd.read_csv("biketest.csv", parse_dates = ["datetime"])
test.head()
```

```python
train["weekday"]=train['datetime'].dt.dayofweek # year - month - day - hour 
train["month"]=train['datetime'].dt.month
train["year"]=train['datetime'].dt.year
train["hour"]=train['datetime'].dt.hour
train.head()
#1 월 2 화 3 수 ...
```

```python
train["tempe"]=train["temp"]+train["atemp"]
train.head()
```

```python
test["weekday"]=test['datetime'].dt.dayofweek # year - month - day - hour 
test["month"]=test['datetime'].dt.month
test["year"]=test['datetime'].dt.year
test["hour"]=test['datetime'].dt.hour
test.head()
```

```python
test["tempe"]=test["temp"]+test["atemp"]
test.head()
```

## Train

```python
feature_names = ["holiday","weather","hour","atemp","year","temp","month"]

feature_names
```

```python
X_train = train[feature_names]
X_train.head()
```

```python
X_test = test[feature_names]
X_test.head()
```

```python
y_train = train["count"]
y_train.head()
```

```python
# from sklearn.tree import DecisionTreeRegressor
# model = DecisionTreeRegressor(random_state = 27)
# model

from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor (n_estimators=1000,random_state =1, max_depth=15, n_jobs=-1)
model
```

# Score

```python
from sklearn.metrics import make_scorer
def rmsle(predict, actual):
    predict=np.array(predict)
    actual=np.array(actual)
    
    log_predict=np.log(predict+1)
    log_actual=np.log(actual+1)
    
    difference = log_predict - log_actual
    square_difference = difference**2
    mean_square_difference = square_difference.mean()
    
    score = np.sqrt(mean_square_difference)
    
    return score

rmsle_score = make_scorer(rmsle)
rmsle_score
```

```python
from sklearn.cross_validation import cross_val_score
score = cross_val_score(model, X_train, y_train, cv=50,scoring=rmsle_score).mean()
print(score)
```
